However, we know there are '?' in the 'stalk-root' features which is used to mention null value (According to the dataset description Link to source - https://archive.ics.uci.edu/dataset/73/mushroom). So we need to impute null values

```python
colname = []
colcodes = []
cols = df.columns
for i in cols:
    colname.append(i)
    colcodes.append(set(df[i].tolist()))
coldir = pd.DataFrame(colname,colcodes)
coldir
```

Below table also shows there is '?' in stalk-root feature. So we have to handle missing values first.

```python
#get value counts in "stock-root" column
print(df['stalk-root'].value_counts())
```

## ---- Data Preprocessing starts---

```python
# Handling missing value
# According to below values, there are more than 2000 '?' in 'stalk-root'
# So we cannot remove those values.
# Since these are categorical data, we can choose mode imputation
```

```python
# change '?' to mode of feature
modes = df['stalk-root'].mode()
df['stalk-root'] = np.where(df['stalk-root']=='?',modes,df['stalk-root'])
```

```python
print(df['stalk-root'].value_counts())
```

```python
df['class'].value_counts()
```

Here, the target feature is 'class' and there are binary values in that feature. So that, need to find the count of each attribute

Here -> edible=e, poisonous=p
Since there are approximately equal number of two values, this can be determined as balance class

Since these are categorical values, need to convert it to relevant numeric values.

```python
# Encode categorical features into numerical features
## try changing the encoder to convert categorical to numerical
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for feature in df.columns:
    df[feature] = le.fit_transform(df[feature])
```

```python
df.head(15)
```

From this stage, need to seperate dataset as X(independent) and y(dependent). Since we know, target feature(y) is 'class' and others are X

```python
## set features to X variable
X = df.drop('class', axis =1)
X.head(10)
```

```python
##get class to one variable as y
y = df['class']
y.head()
```

##---- preprocessing ends ---

Split dataset for train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=42) ##training set 75%
# Here test data set size is 25% from the whole dataset whereas 75% from the dataset are for training set.
# Random state is used to that if we set random_state to a specific value (here 42), the ML algorithm will always generate the same random numbers, regardless of how many times
```

```python
len(X) # number of instances of X (independent)
```

```python
len(X_train) # number of instances of X_train (X part of train dataset)
```

```python
len(X_test) # number of instances of X_test (X part of test dataset)
```

```python
# Scale the features by using MinMax Scaler. Since this problem is binary classification, we want to scale values between 0-1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

Since this is a classification problem, we can use LogisticRegression,RidgeClassifier, DecisionTreeClassifier,GaussianNB,MLPClassifier and RandomForestClassifier algorithms. So that, we can find accuracies of each model and then can find the best model for training dataset

```python
# Import Logistic Regression, Ridge Classifier, Decision Tree
from sklearn.linear_model import LogisticRegression,RidgeClassifier
from sklearn.tree import DecisionTreeClassifier

# Create an object using the LogisticRegression() class
logistic_classifier_model = LogisticRegression()

# Create an object using the RidgeClassifier() class
ridge_classifier_model = RidgeClassifier()

# Create an object using the DecisionTreeClassifier() class
decision_tree_model = DecisionTreeClassifier()
```

```python

# Gaussian Naive Bayes, MLP Classifier and Random Forest models

from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier

# Create an object using the GaussianNB() class
naive_bayes_model = GaussianNB()

# Create an object using the MLPClassifier() class
neural_network_model = MLPClassifier()

# Create Random Forest Classifier object
random_forest_model = RandomForestClassifier()
```

After that, we can train dataset by using above models

```python
# Train the Logistic Classifier model
logistic_classifier_model.fit(X_train_scaled ,y_train)

# Train the Ridge Classifier model
ridge_classifier_model.fit(X_train_scaled ,y_train)

# Train the Decision Tree model
decision_tree_model.fit(X_train_scaled ,y_train)
```

```python
# Train the Naive Bayes model
naive_bayes_model.fit(X_train_scaled ,y_train)

# Train the Neural Network model
neural_network_model.fit(X_train_scaled ,y_train)

# Train the Random Forest model
random_forest_model.fit(X_train_scaled, y_train)
```

Here we can do the prediction by using above models and X_test dataset

```python
# Make prediction using the test dataset on Logistic Classifier model
logistic_pred = logistic_classifier_model.predict(X_test)

# Make prediction using the test dataset on Ridge Classifier model
ridge_pred = ridge_classifier_model.predict(X_test)

# Make prediction using the test dataset on Decision Tree model
tree_pred = decision_tree_model.predict(X_test)
```

```python
# Make prediction using the test dataset on Naive Bayes model
naive_bayes_pred = naive_bayes_model.predict(X_test)

# Make prediction using the test dataset on Neural Network model
neural_network_pred = neural_network_model.predict(X_test)

# Make prediction using the test dataset on Random Forest model
random_forest_pred = random_forest_model.predict(X_test)
```

Generating reports for retrive performance level of each model

```python
from sklearn.metrics import classification_report
# Create a Classification Report for Logistic Classifier model
logistic_report = classification_report(y_test, logistic_pred,zero_division=1)

# Create a Classification Report for Ridge Classifier model
ridge_report = classification_report(y_test, ridge_pred,zero_division=1)

# Create a Classification Report for Decision Tree model
tree_report = classification_report(y_test, tree_pred,zero_division=1)

# Create a Classification Report for Naive Bayes model
naive_bayes_report = classification_report(y_test, naive_bayes_pred,zero_division=1)

# Create a Classification Report for Neural Network model
neural_network_report = classification_report(y_test, neural_network_pred,zero_division=1)

# Create a classification Report for Random Forest model
random_forest_report = classification_report(y_test, random_forest_pred,zero_division=1)
```

```python
# Print the report of the Logistic Regression model
print('***** Logistic Regression *****')
print(logistic_report)

# Print the report of the Ridge Regression model
print('***** Ridge Regression *****')
print(ridge_report)

# Print the report of the Decision Tree model
print('***** Decision Tree *****')
print(tree_report)

# Print the report of the Naive Bayes model
print('***** Naive Bayes *****')
print(naive_bayes_report)

# Print the report of the Neural Network model
print('***** Neural Network *****')
print(neural_network_report)

# Print the report of Random Forest model
print('***** Random Forest *****')
print(random_forest_report)
```

Since these performance levels from above models, we can choose Decision Tree for model training

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(
    decision_tree_model ,
    X_test,
    y_test
);
```

```python
# Evaluate the model on the test set
y_pred = decision_tree_model.predict(X_test) ##predict data from known class of testing set
from sklearn.metrics import accuracy_score
# Calculate the accuracy using score
accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy of using decision tree to the test dataset
print(accuracy)
```
